# 01 — Preprocessing
Clean, encode, scale, and split the dataset.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
# Load raw dataset
df = pd.read_csv('../data/plant_health_data.csv')
df.head()
```

```python
# Drop non-ML columns
drop_cols = [col for col in ['Timestamp', 'Plant_ID'] if col in df.columns]
df = df.drop(columns=drop_cols)
df.head()
```

```python
# Handle missing values (drop rows with any missing)
df = df.dropna()
df.isnull().sum()
```

```python
# Encode target label
label_map = {'Healthy': 0, 'Moderate Stress': 1, 'High Stress': 2}
if 'Plant_Health_Status' in df.columns:
    df['Plant_Health_Status'] = df['Plant_Health_Status'].map(label_map)
else:
    raise ValueError('Plant_Health_Status column not found in dataset.')
df['Plant_Health_Status'].value_counts()
```

```python
# Feature/target split
X = df.drop('Plant_Health_Status', axis=1)
y = df['Plant_Health_Status']
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
# Save processed dataset
processed = pd.DataFrame(X, columns=X.columns)
processed['Health_Status'] = y.values
processed.to_csv('../data/processed_dataset.csv', index=False)
print('Processed dataset saved.')
```

# 06 — Predict New Plant Readings
Load a saved model and scaler, input new sensor readings, and predict plant health status.

```python
import joblib
import os
import numpy as np
import pandas as pd
# Load scaler and model (choose RF or SVM)
scaler = joblib.load('../../models/scaler.pkl') if os.path.exists('../../models/scaler.pkl') else None
model = joblib.load('../../models/random_forest_model.pkl')  # or svm_model.pkl
# Input new readings (array of dicts)
input_dicts = [
    {
        'Soil_Moisture': 32.5,
        'Ambient_Temperature': 27.8,
        'Soil_Temperature': 25.3,
        'Humidity': 65.2,
        'Light_Intensity': 1200,
        'Soil_pH': 6.8,
        'Nitrogen_Level': 45,
        'Phosphorus_Level': 18,
        'Potassium_Level': 38,
        'Chlorophyll_Content': 42.7,
        'Electrochemical_Signal': 0.85
    },
    {
        'Soil_Moisture': 18.0,
        'Ambient_Temperature': 34.2,
        'Soil_Temperature': 31.1,
        'Humidity': 40.5,
        'Light_Intensity': 900,
        'Soil_pH': 5.9,
        'Nitrogen_Level': 22,
        'Phosphorus_Level': 10,
        'Potassium_Level': 20,
        'Chlorophyll_Content': 25.3,
        'Electrochemical_Signal': 0.42
    },
    {
        'Soil_Moisture': 10.2,
        'Ambient_Temperature': 38.5,
        'Soil_Temperature': 35.0,
        'Humidity': 28.7,
        'Light_Intensity': 700,
        'Soil_pH': 5.2,
        'Nitrogen_Level': 10,
        'Phosphorus_Level': 5,
        'Potassium_Level': 8,
        'Chlorophyll_Content': 12.1,
        'Electrochemical_Signal': 0.21
    }
    # Add more dicts as needed
 ]
# Convert to DataFrame
X_new = pd.DataFrame(input_dicts)
if scaler is not None:
    X_new_scaled = scaler.transform(X_new)
else:
    X_new_scaled = X_new.values
preds = model.predict(X_new_scaled)
label_map = {0: 'Healthy', 1: 'Moderate Stress', 2: 'High Stress'}
for i, pred in enumerate(preds):
    print(f'Sample {i+1} - Predicted Health Status:', label_map[pred])
```
